```python
from sklearn.datasets import load_iris
from sklearn.cluster import MiniBatchKMeans, KMeans
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = load_iris()
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
print(df)
```

```python
features = data.data
labels = data.target
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

features_to_plot = [1,3]

plt.scatter(features[:,features_to_plot[0]], features[:,features_to_plot[1]], s=2**7, 
            c = labels, edgecolor = 'white', alpha= 0.85, cmap='autumn')
plt.grid()
plt.xlabel(data.feature_names[features_to_plot[0]])
plt.xlabel(data.feature_names[features_to_plot[1]])
```

```python
# ALTHOUGH KMEANS IS USING ONLY FEATURES IN THIS DATASET THEY ARE 
# LABELED SO YOU SEE THE CORRECTNESS

# TRY LESS OR MORE CLASSES (I.E 2 OR 4). IF DATA WASN'T LABELED YOU WOULDN'T RECOGNIZE THAT IN REALITY 
# VERSICOLOR AND VIRGINICA ARE TWO DIFFERENT TYPES OF IRIS.

nr_of_classes = 3

k_means = KMeans(n_clusters = nr_of_classes, init = 'k-means++', 
                 max_iter = 999, n_init = 1, random_state = 101)
mb_k_means = MiniBatchKMeans(n_clusters = nr_of_classes, init = 'k-means++', 
                 max_iter = 999, n_init = 1, random_state = 101)

k_means.fit(features)
mb_k_means.fit(features)
```

```python
# FROM 0 TO 3 WITH THEIR RESPECTIVE CENTROID CENTERS
features_to_plot = [1,2]

plt.scatter(features[:,features_to_plot[0]], features[:,features_to_plot[1]], s=2**7, 
            c = labels, edgecolor = 'white', alpha= 0.85, cmap='autumn')
plt.grid()
plt.xlabel(data.feature_names[features_to_plot[0]])
plt.xlabel(data.feature_names[features_to_plot[1]])

plt.scatter(k_means.cluster_centers_[:,features_to_plot[0]], 
            k_means.cluster_centers_[:,features_to_plot[1]], 
            s = 2**6, marker = 's', c = 'blue')
plt.show()
```

```python
print(k_means.cluster_centers_)

# 3 ROWS = 3 CLASSES
# 4 COLUMNS REPRESENT THE VALUE IN EACH FEATURE SPACE (4 DIMENSIONS)
```

```python
features_to_plot = [1,2]

# COLORS SET TO KMEANS CALCULATED LABELS
plt.scatter(features[:,features_to_plot[0]], features[:,features_to_plot[1]], s=2**7, 
            c = k_means.labels_, edgecolor = 'white', alpha= 0.85, cmap='autumn')
plt.grid()
plt.xlabel(data.feature_names[features_to_plot[0]])
plt.xlabel(data.feature_names[features_to_plot[1]])

plt.scatter(k_means.cluster_centers_[:,features_to_plot[0]], 
            k_means.cluster_centers_[:,features_to_plot[1]], 
            s = 2**6, marker = 's', c = 'blue')
plt.show()
```
