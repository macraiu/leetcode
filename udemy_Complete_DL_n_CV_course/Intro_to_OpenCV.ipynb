```python
import cv2
#cv2.__version__
import matplotlib.pyplot as plt
import os 

cwd = os.getcwd()
img = cv2.imread(cwd + "/imgs/dog.jpg") 
# specify a 0 as a second argument for importing it in grey scale
# default set to 1
```

```python
def CVImShow(img, title = "Default"):
    cv2.imshow(title, img)
    # Window name, and variable

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def CVImSave(img, title):
    cv2.imwrite(title, img)
```

## Change color spaces

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
# Used for segmentation mainly
CVImShow(hsv_img)
```

```python
plt.imshow(hsv_img)
```

```python
grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# grey scale
CVImShow(grey_img)
```

```python
plt.imshow(grey_img)
```
