## Drawing with CV2
### Add a line

```python
import cv2
#cv2.__version__
import matplotlib.pyplot as plt
import os 

cwd = os.getcwd()
img = cv2.imread(cwd + "/imgs/dog.jpg") 
# specify a 0 as a second argument for importing it in grey scale
# default set to 1
```

```python
def loadDog():
    return cv2.imread(cwd + "/imgs/dog.jpg") 
```

```python
start_point=(0,0)
end_point=(400, 400)
color=(255,0,0)
thickness=10

_ = cv2.line(img, start_point, end_point, color, thickness)
```

```python
cv2.imshow("Default", img)
# Window name, and variable

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img)
# WHEN MATPLOTLIB SHOWS THE IMAGE THE COLOR OF THE LINE WITH BE DEFINED AS RGB
# INSTEAD IN CV2 IMSHOW THE COLOR WILL BE IN BGR
```

```python
img = loadDog()

circleCenter = (100,50)
circleRadius = 10
colorBGR = (255,0,0)
thickness = 3

cv2.circle(img, circleCenter, circleRadius, colorBGR, thickness)

plt.imshow(img)
```

```python
img = loadDog()

textContent = "A dog"
textPosition = (50, 200)
textFont = cv2.FONT_HERSHEY_COMPLEX
textScale = 1
textColor = (0, 255, 0)
textThickness = 2

cv2.putText(img, textContent, textPosition, textFont, textScale, textColor, textThickness)

plt.imshow(img[:, :, ::-1]) # cv stores the dog as BGR, so retrieve the img channels in reverse order.
```
