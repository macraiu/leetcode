## Geometrical Transformations 
### Scaling

```python
import cv2
#cv2.__version__
import matplotlib.pyplot as plt
import os 
import numpy as np
cwd = os.getcwd()
img = cv2.imread(cwd + "/imgs/dog.jpg") 
# specify a 0 as a second argument for importing it in grey scale
# default set to 1
```

```python
def loadDog():
    return cv2.imread(cwd + "/imgs/dog.jpg") 
```

```python
#img = reloadDog()
scaledDog = cv2.resize(img, (50,50)) # resizes to these dimensions
# RESIZE used in DL for reducing the computational power
plt.imshow(scaledDog)
```

### Translation

```python
img = loadDog()
rows, cols, channel = img.shape
M = np.float32([[1,0,100],[0,1,50]])
translate_image = cv2.warpAffine(img, M, (cols, rows))
plt.imshow(translate_image)
```

### Rotate

```python
M = cv2.getRotationMatrix2D((cols/2, rows/2), 56, 1) #center of rotation
rotatedImage = cv2.warpAffine(img, M, (cols, rows))
plt.imshow(rotatedImage)
```

### Flip

```python
flippedImage = cv2.flip(img, 1)
plt.imshow(flippedImage)
```

## Affine vs. Perspective Transformations
### Affine transformations: parallel lines remmain parallel. As opposed to perspective transformations

```python
# AFFINE
rows, cols, channel = img.shape

# 3 POINTS
pts1 = np.float32([[50,50], [200,50], [50,200]]) # Starting points
pts2 = np.float32([[50,90], [150,100], [100,220]]) # Ending points

M = cv2.getAffineTransform(pts1, pts2)

affineTFImage = cv2.warpAffine(img, M, (cols, rows))

plt.imshow(affineTFImage)
```

```python
# PERSPECTIVE
pts1 = np.float32([[400,400], [268,52], [28,387], [300,300]])
pts2 = np.float32([[0,0], [600,150], [0,1000], [1000,1000]])

M = cv2.getPerspectiveTransform(pts1, pts2)

perspectiveTFImage = cv2.warpPerspective(img, M, (500, 650))

plt.imshow(perspectiveTFImage)
```
